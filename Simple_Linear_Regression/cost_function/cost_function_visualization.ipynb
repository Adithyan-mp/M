```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.DataFrame([[1,2],[2,4]])
x=data[0]
y=data[1]
```

```python
w = 2
b = 0
```

```python
def hypothesis(w,b,x)->np.array:
    temp=np.zeros(len(x))
    for i in range(len(x)):
        temp[i]=w*x[i]+b
    return temp
```

```python
y_pred=hypothesis(w,b,x)
plt.plot(x, y, 'x', c='r')  
plt.plot(x, y_pred)  

plt.xlabel('x')
plt.ylabel('y')
plt.title('Cost function wrt W')
plt.show()
```

```python
def J(w, b):
    m = len(x)
    y_pred = hypothesis(w, b, x)
    return 1 / (2 * m) * np.sum((y_pred - y) ** 2)
```

```python
W = np.arange(-3, 7, 0.1)

J_values = [J(w, 0) for w in W]

plt.plot(W, J_values)
plt.xlabel('w')
plt.ylabel('J(w, b)')
plt.title('Cost function J(w, b) with respect to w')
plt.xticks(np.arange(-3,7,1))
plt.grid(True)
plt.show()
```
