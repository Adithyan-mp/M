```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data initialization
x = np.array([1.0, 2.0])
y = np.array([300, 500])
```

```python
# Calculate the number of training sets
m = len(x)
print(f"the number of training set is {m}")
```

```python
# Print specific training set
i = 1
x_i = x[i]
y_i = y[i]
print(f"(x^{i}, y^{i}) = ({x_i}, {y_i})")
```

```python
# Scatter plot visualization
plt.scatter(x, y, marker="x", c="r")
plt.xlabel("size")
plt.ylabel("prize")
plt.title("House price prediction")
plt.show()
```

```python
# Initialize model parameters
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
# Define the linear model function
def calculate_model_output(w, x, b):
    m = np.shape(x)
    f_w_b = np.zeros(m)
    for i in range(len(x)):
        f_w_b[i] = (w * x[i]) + b
    return f_w_b
```

```python
# Calculate model output
temp_f_w_b = calculate_model_output(w, x, b)

# Plot model prediction
plt.plot(x, temp_f_w_b, c='black', label='our prediction')
plt.scatter(x, y, marker="x", c="r", label='Actual values')
plt.xlabel("size")
plt.ylabel("prize")
plt.yticks(range(100, 600, 100))
plt.title("House price prediction")
plt.legend()
plt.show()
```
